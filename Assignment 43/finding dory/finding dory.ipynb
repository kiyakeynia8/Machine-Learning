```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

from KNN_main import KNN
```

```python
maman_dory = cv2.imread("maman_dory.jpg")
maman_dory = cv2.cvtColor(maman_dory, cv2.COLOR_BGR2RGB)
maman_dory = cv2.resize(maman_dory, (0, 0), fx= 0.5, fy=0.5)
print(maman_dory.shape)
plt.imshow(maman_dory)
plt.show()
```

```python
pixels_list_rgb = maman_dory.reshape(-1, 3)

pixels_list_rgb.shape

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection = "3d")

ax.scatter(pixels_list_rgb[:, 0], pixels_list_rgb[:, 1], pixels_list_rgb[:, 2], c = pixels_list_rgb/255)
plt.show()
```

```python
maman_dory_hsv = cv2.cvtColor(maman_dory, cv2.COLOR_RGB2HSV)
```

```python
pixels_list_hsv = maman_dory.reshape(-1, 3)

pixels_list_hsv.shape

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection = "3d")

ax.scatter(pixels_list_hsv[:, 0], pixels_list_hsv[:, 1], pixels_list_hsv[:, 2], c=pixels_list_rgb/255)
plt.show()
```

```python
light_blue = (0, 0, 100)
dark_blue = (55, 155, 255)

mask = cv2.inRange(maman_dory, light_blue, dark_blue)
plt.imshow(mask, cmap="gray")
plt.show()
```

```python
light_yellow = (150, 130, 30)
dark_yellow = (255, 255, 95)

mask_white = cv2.inRange(maman_dory, light_yellow, dark_yellow)
plt.imshow(mask_white, cmap="gray")
plt.show()
```

```python
light_black = (0, 0, 40)
dark_black = (30, 40, 100)

mask_black = cv2.inRange(maman_dory, light_black, dark_black)
plt.imshow(mask_black, cmap="gray")
plt.show()
```

```python
final_mask = mask + mask_white + mask_black
plt.imshow(final_mask, cmap="gray")
plt.show()
```

```python
final_result = cv2.bitwise_and(maman_dory, maman_dory, mask = final_mask)
plt.imshow(final_result)
plt.show()
```

```python
X_train = pixels_list_hsv / 255
X_train
```

```python
Y_train = final_mask.reshape(-1,) // 255
Y_train
```

```python
knn = KNN(3)
knn.fit(X_train, Y_train)
```

```python
new_pixel = np.array([33, 62, 241])/255
knn.predict([new_pixel])
```

```python
baba_dory = cv2.imread("baba_dory.jpg")
baba_dory = cv2.cvtColor(baba_dory, cv2.COLOR_BGR2RGB)
baba_dory = cv2.resize(baba_dory, (0, 0), fx= 0.5, fy=0.5)
print(baba_dory.shape)
plt.imshow(baba_dory)
plt.show()
```

```python
pixels_list_hsv = baba_dory.reshape(-1, 3)
X_test = pixels_list_hsv / 255
Y_pred = knn.predict(X_test)
Y_pred
```

```python
output = np.array(Y_pred).reshape(baba_dory.shape[:2])
plt.imshow(output, cmap="gray")
plt.show()
```

```python
output = output.astype("uint8")
```

```python
final_result = cv2.bitwise_and(baba_dory,baba_dory, mask=output)
plt.imshow(final_result)
plt.show()
```
